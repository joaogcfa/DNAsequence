### Objetivo da entrega:

Com duas sequências de nucleotídeos de DNA a[i] e b[j], de tamanhos n e m respectivamente,pesos wmat, wmis e wgap para match, mismatch e gap respectivamente encontrar da seguinte maneira os pares mais semlhantes:

1. Gerar uma subsequencia sb=b[j,j+1,...,j+k] de b, de tamanho aleatório k, 1<=k<=m, e 0<=j<=m 
2. Gerar um número aleatório inteiro positivo p
3. Gerar p subsequencias sa=a[i,i+1,...,i+k] de a, com tamanho k calculado no passo (1), 0<=i<=n
4. Calcular os scores de cada par (sa,sb) com os pesos wmat, wmis e wgap
5. Devolver o score máximo m entre os scores do passo (4) e as subsequencias associadas a ele

A intuição do projeto será, portanto:

Implementar um programa C++ para ler um arquivo contendo os tamanhos de duas sequencias de DNA, seguidos das duas sequencias, uma por linha. Calcular o score máximo utilizando o algoritmo acima, assim como as subsequencias associadas a ele.

# Metadologia

### Gerar arquivo de entrada:

````python
import random
n = 10 # tamanho da primeira sequência
m = 40 # tamanho da segunda sequência
file = 'dna.seq' # nome do arquivo a ser gerado
f = open(file, 'w')
seq=[str(n)+'\n',
     str(m)+'\n',
     ''.join(random.choices(['A','T','C','G','-'],k=n))+'\n',
     ''.join(random.choices(['A','T','C','G','-'],k=m))]
f.writelines(seq)
f.close()
print(''.join(seq))
````

### Com o arquivo de entrada podemos rodar o programa

````python
import subprocess
import time

with open('dna.seq') as f:
    start = time.perf_counter()
    proc = subprocess.run(['./busca_local'], input=f.read(), text=True, capture_output=True)
    end = time.perf_counter()

    print('Saída:\n', proc.stdout)
    print('Stderr:', proc.stderr)
    print('Tempo total(s):', end - start)
````

### Para gerar um arquivo txt é necessário rodar a seguinte linha de comando no terminal:

 ```./main < dna.seq > out.txt```
